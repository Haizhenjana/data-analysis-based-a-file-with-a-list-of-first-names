```python
import csv

#Die relevante Daten(von Spalte count) des bestimmte Vornamen von jedes Jahr in einer Liste speichern
name = "Anna"
gender = "F"
state = "AK"
xs = []
ys = []
with open("names80until87.csv",newline='') as file:
    csv_file=csv.reader(file,delimiter=",")
    for line in csv_file:
        if line[1] == name and line[3] == gender and line[4] == state:
            xs.append(line[2])
            ys.append(int(line[5]))

#Die Daten visualisieren mit Modul mitplotlib
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(xs, ys, color = "#00ff05", linestyle = "dashed", marker = "o")
plt.show()
            
```
